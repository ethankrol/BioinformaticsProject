```python
import pandas as pd
from sklearn.cluster import AffinityPropagation
from sklearn.preprocessing import StandardScaler
import os

# Define the results directory
results_dir = "results_assignment3"

# Ensure the results directory exists
os.makedirs(results_dir, exist_ok=True)

# List of gene counts to iterate over
num_genes_list = [10, 100, 1000, 5000, 10000]

clusters = {}

# Loop through each specified number of genes
for num_genes in num_genes_list:
    # Load gene expression data for the top 'num_genes' variable genes
    gene_data_path = f"{results_dir}/top_{num_genes}_variable_genes.tsv"
    df = pd.read_csv(gene_data_path, sep='\t', index_col=0, header=0)
    
    # Load the corresponding metadata
    metadata_path = "results/filtered_metadata.tsv"
    metadata = pd.read_csv(metadata_path, sep='\t', index_col=0, header=0)
    
    # Transpose the data so that each row represents a sample
    df = df.T
    
    # Scale the data to have mean 0 and variance 1
    scaler = StandardScaler()
    df_scaled = scaler.fit_transform(df)
    
    # Perform Affinity Propagation clustering
    affinity = AffinityPropagation(random_state=1)
    affinity.fit(df_scaled)
    
    # Retrieve cluster labels
    clusters = affinity.labels_
    
    # Determine the number of clusters found
    num_clusters = len(set(clusters))
    
    # Create a DataFrame with cluster assignments
    clusters_df = pd.DataFrame({
        'refinebio_accession_code': df.index,
        'Cluster': clusters
    })
    
    # Merge cluster information with metadata
    metadata_with_clusters = metadata.merge(clusters_df, on='refinebio_accession_code')
    
    # Define the output filename with the determined number of clusters
    output_filename = f"{num_clusters}_affinity_propagation_clusters_from_{num_genes}_genes.tsv"
    output_path = os.path.join(results_dir, output_filename)
    
    # Save the merged metadata with cluster assignments to a TSV file
    metadata_with_clusters.to_csv(output_path, sep='\t', index=False)
    
    # Optional: Print a summary of the clustering results
    print(f"Processed {num_genes} genes: Found {num_clusters} clusters.")
    clusters[num_genes] = num_clusters
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

num_genes_list = [10, 100, 1000, 5000, 10000]

###################################################
## Chi-Squared Results
###################################################

comparison_results = [[None] * (len(num_genes_list)+1) for _ in range(len(num_genes_list)+1)]

for i, num_genes in enumerate(num_genes_list):
    comparison_results[i+1][0] = num_genes
    comparison_results[0][i+1] = num_genes

comparison_results[0][0] = 'Num Genes'


# sample each pair from num_genes_list
for i in range(len(num_genes_list)):
    for j in range(i+1, len(num_genes_list)):
        num_genes1 = num_genes_list[i]
        num_genes2 = num_genes_list[j]
        
        # read the data
        df1 = pd.read_csv(f"results_assignment3/affinity_propagation_clusters_from_{num_genes1}_genes.tsv", sep='\t')
        df2 = pd.read_csv(f"results_assignment3/affinity_propagation_clusters_from_{num_genes2}_genes.tsv", sep='\t')

        # contingency table
        contingency_table = pd.crosstab(df1['Cluster'], df2['Cluster'])

        # Perform the Chi-Square test of independence
        chi2, p_value, dof, expected = chi2_contingency(contingency_table)

        p_value = f'{float(p_value):.3e}'

        # add results to table
        comparison_results[i+1][j+1] = p_value


# Create a figure and axis
fig, ax = plt.subplots()

# Hide axes
ax.axis('tight')
ax.axis('off')

# Create a table and add it to the axis
table = ax.table(cellText=comparison_results, loc='center', cellLoc='center')

table.auto_set_font_size(False)
table.set_fontsize(13)
table.scale(1.5, 1.5) 

plt.title('Comparison of Affinity Propagation Clustering Results\nwith Different Numbers of Genes Used in Clustering', fontsize=16, y=0.8)

caption = "This table displays the p-values obtained from pairwise Chi-Squared tests."
plt.text(0, -0.03, caption, ha='center', va='top', fontsize=12, wrap=True)

# Save the table as a PNG file
plt.savefig('results_assignment3/affinity_propagation_num_genes_comparison.png', bbox_inches='tight', dpi=300)

###################################################
## Sankey Diagram
###################################################

import plotly.graph_objects as go

# Initialize lists to store source, target, value, and labels
source = []
target = []
value = []
labels = []
label_map = {}  # Mapping from (num_genes, cluster) to label index
label_index = 0

# Create labels and links for Sankey diagram
for i in range(len(num_genes_list) - 1):
    num_genes1 = num_genes_list[i]
    num_genes2 = num_genes_list[i + 1]

    # Read the data for the two gene subsets
    df1 = pd.read_csv(f"results_assignment3/affinity_propagation_clusters_from_{num_genes1}_genes.tsv", sep='\t')
    df2 = pd.read_csv(f"results_assignment3/affinity_propagation_clusters_from_{num_genes2}_genes.tsv", sep='\t')

    # Merge the two dataframes on the sample identifier
    merged_df = pd.merge(df1, df2, on="refinebio_accession_code", suffixes=(f"_{num_genes1}", f"_{num_genes2}"))

    # Get the unique clusters for each gene set
    unique_clusters1 = sorted(merged_df[f'Cluster_{num_genes1}'].unique())
    unique_clusters2 = sorted(merged_df[f'Cluster_{num_genes2}'].unique())

    # Generate cluster labels dynamically for each gene set
    for cluster in unique_clusters1:
        label = f"{num_genes1} genes<br>Cluster {cluster}"
        if (num_genes1, cluster) not in label_map:
            label_map[(num_genes1, cluster)] = label_index
            labels.append(label)
            label_index += 1

    for cluster in unique_clusters2:
        label = f"{num_genes2} genes<br>Cluster {cluster}"
        if (num_genes2, cluster) not in label_map:
            label_map[(num_genes2, cluster)] = label_index
            labels.append(label)
            label_index += 1

    # Create contingency table and map the source and target links
    contingency_table = pd.crosstab(merged_df[f'Cluster_{num_genes1}'], merged_df[f'Cluster_{num_genes2}'])

    for cluster1 in contingency_table.index:
        for cluster2 in contingency_table.columns:
            count = contingency_table.loc[cluster1, cluster2]
            if count > 0:
                source.append(label_map[(num_genes1, cluster1)])
                target.append(label_map[(num_genes2, cluster2)])
                value.append(count)

# Create the Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color="blue"
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
)])

fig.update_layout(title_text="Sankey Diagram of Affinity Propagation Clustering Results\nwith Different Numbers of Genes Used in Clustering", font_size=10)

# Save or display the diagram
fig.write_image("results_assignment3/affinity_propagation_num_genes_sankey_corrected.png")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import numpy as np

# Load expression data
filtered_expression_df = pd.read_csv('results/filtered_data.tsv', sep='\t', index_col='Symbol')

# Calculate variance for each gene and select top 5,000 variable genes
gene_variances = filtered_expression_df.var(axis=1)
top_5000_genes = gene_variances.nlargest(5000).index
top_5000_gene_data = filtered_expression_df.loc[top_5000_genes]

# Transpose the data for clustering (rows are samples, columns are genes)
top_5000_gene_data_T = top_5000_gene_data.T

# Standardize the data to mean 0 and variance 1 (Z-score normalization)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(top_5000_gene_data_T)
data_scaled_df = pd.DataFrame(data_scaled, index=top_5000_gene_data_T.index, columns=top_5000_gene_data_T.columns)

# Perform hierarchical clustering (for dendrograms)
row_linkage = sch.linkage(data_scaled, method='average')
col_linkage = sch.linkage(data_scaled.T, method='average')

# Perform clustering using AgglomerativeClustering for annotation
agg_clustering = AgglomerativeClustering(n_clusters=4)  # You can adjust the number of clusters
clusters_5000 = agg_clustering.fit_predict(data_scaled)
data_scaled_df['Cluster_5000_genes'] = clusters_5000

# Load metadata
metadata = pd.read_csv('results/filtered_metadata.tsv', sep='\t', index_col=0)

# Merge metadata with cluster information for annotation
annotation_data = pd.DataFrame({
    'Cluster_5000_genes': data_scaled_df['Cluster_5000_genes'],
    'Sample_Group': metadata['time_status']
})

# Generate the heatmap
sns.set(style="white")

# Create a clustermap (heatmap + dendrograms)
g = sns.clustermap(
    data_scaled_df.drop('Cluster_5000_genes', axis=1),  # Drop cluster information from the actual data
    row_cluster=True,
    col_cluster=True,
    row_linkage=row_linkage,
    col_linkage=col_linkage,
    figsize=(12, 10),
    cmap='vlag',  # Color map
    annot=False,
    xticklabels=True,  # Show gene names on the columns
    yticklabels=False,  # Hide sample names
    row_colors=annotation_data['Cluster_5000_genes'].map({0: 'red', 1: 'blue', 2: 'green', 3: 'purple'}),
    col_colors=annotation_data['Sample_Group'].map({'Group1': 'red', 'Group2': 'blue'})  # Add sample group annotation
)

# Add a title and save the heatmap
g.fig.suptitle('Heatmap of 5,000 Most Variable Genes with Clustering Annotations', y=1.02)
plt.savefig('results_assignment3/heatmap_5000_genes_clustering.png', dpi=300, bbox_inches='tight')
plt.show()
```
